```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

```python
mlp_data = pd.read_csv('gr_data.txt', delimiter = ',')
```

```python
mlp_data
```

```python
sorted_data = mlp_data.sort_values(by = 'accuracy', ascending = False)
```

```python
sorted_data
```

```python
sorted_data.dtypes
```

```python
sorted_data.rename(columns = {'H.L.N':'HLN'}, inplace = True)
```

```python
sorted_data.dtypes
```

```python
sorted_data = sorted_data[sorted_data.HLN !=  30]
sorted_data
```

```python
plt.scatter(sorted_data['H.L.N.'], sorted_data['accuracy'], alpha=0.5, color = 'blue', cmap='inferno')
pyplot.title("Correlation between hidden layers sizes and accuracy of a NN, range 10 : 600", fontsize = 13)
plt.xlabel('H.L.S', fontsize = 13)
plt.ylabel('Accuracy', fontsize = 13)
```

```python
plt.scatter(sorted_data['H.L.N.'], sorted_data['work time'], alpha=0.5, color = 'green', cmap='inferno')
pyplot.title("Correlation between hidden layers sizes and work time(seconds) of a NN, range 10 : 600", fontsize = 13)
plt.xlabel('H.L.S', fontsize = 13)
plt.ylabel('WT', fontsize = 13)
```

```python
plt.scatter(sorted_data['H.L.N.'], sorted_data['loss'], alpha=0.7, color = 'red', cmap='inferno')
pyplot.title("Correlation between hidden layers sizes and loss of a NN, range 10 : 600", fontsize = 13)
plt.xlabel('H.L.S', fontsize = 13)
plt.ylabel('loss', fontsize = 13)
```

```python
relu_data = pd.read_csv('relu_data.txt', delimiter = ',')
```

```python
relu_data
```

```python
relu_data.rename(columns = {'H.L.N.' : 'HLN'}, inplace = True)
```

```python
tanh_data = sorted_data.sort_values(by = 'HLN', ascending = False)
tanh_data = tanh_data[(sorted_data.HLN == 450) | (sorted_data.HLN == 150) | (sorted_data.HLN == 300) | (sorted_data.HLN == 390) | (sorted_data.HLN == 30)]

```

```python
tanh_data
```

```python
plt.plot(relu_data['H.L.N.'], relu_data['accuracy'], alpha=0.5, color = 'green', label = 'relu')
plt.plot(tanh_data['H.L.N.'], tanh_data['accuracy'], alpha=0.5, color = 'red', label = 'tanh')
pyplot.title("Difference in accuracy between relu and tanh activation functions", fontsize = 13)
plt.xlabel('H.L.S', fontsize = 13)
plt.ylabel('Accuracy', fontsize = 13)
plt.legend()
```
